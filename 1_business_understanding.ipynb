# Kenya Road Accident Analysis

## Business Understanding

### Project Background

Road traffic accidents are a major public safety concern in Kenya, leading to fatalities, injuries, property damage, and economic losses. Understanding the factors associated with road crashes is essential for improving road safety, informing policy decisions, and prioritizing interventions in high-risk areas.

This project analyzes historical road accident data in Kenya to identify patterns and trends in crash occurrence, severity, and geographic distribution. The analysis focuses on understanding when accidents occur, where they occur, and which road users are most affected.

---

## Problem Statement

Road traffic accidents continue to pose significant social and economic challenges in Kenya. While various road safety initiatives have been implemented, crashes remain common across different regions and road user groups. There is a need to better understand accident patterns, identify high-risk periods and locations, and evaluate the involvement of vulnerable road users such as pedestrians and motorcyclists.

By analyzing historical accident data, this project seeks to generate actionable insights that can support evidence-based road safety interventions and decision-making.

---

## Project Goal

To analyze road accident patterns in Kenya and identify key temporal, geographic, and road-user-related factors associated with crash occurrence and severity.

---

## Project Objectives

The specific objectives of this project are:

1. Analyze road accident trends across different years.
2. Identify seasonal patterns in crash occurrence by month and quarter.
3. Examine accident frequencies by day of the week and hour of the day.
4. Assess the prevalence and distribution of fatality-related crashes.
5. Investigate pedestrian-related crash patterns.
6. Investigate motorcycle-related crash patterns.
7. Evaluate the involvement of matatus in road accidents.
8. Identify geographic accident hotspots using spatial analysis techniques.
9. Develop data-driven recommendations to support road safety initiatives.

---

## Key Business Questions

This analysis seeks to answer the following questions:

### Temporal Analysis

* How have road accident frequencies changed over time?
* Which months and quarters experience the highest number of crashes?
* Which days of the week record the most accidents?
* During which hours of the day are crashes most common?

### Crash Severity Analysis

* What proportion of crashes involve fatalities?
* How has the fatality rate changed over time?

### Vulnerable Road User Analysis

* How common are pedestrian-related crashes?
* How common are motorcycle-related crashes?
* When are pedestrian and motorcycle crashes most likely to occur?

### Public Transport Analysis

* What proportion of crashes involve matatus?
* What temporal patterns are associated with matatu-related crashes?

### Geospatial Analysis

* Which locations experience the highest concentration of crashes?
* Are there identifiable accident hotspots and high-risk road corridors?

---

## Expected Outcomes

The project is expected to:

* Reveal major trends and patterns in road accidents.
* Identify high-risk periods and locations.
* Highlight the involvement of vulnerable road users.
* Support evidence-based road safety planning.
* Provide recommendations for reducing road traffic accidents and fatalities.
