# Kenya Road Accident Analysis

## Problem Statement
Road traffic accidents remain a major public safety concern in Kenya. Understanding when and where accidents occur can help identify accident hotspots and high-risk periods.

## Project Objectives

1. Analyze road accident trends over time.
2. Identify accident hotspots.
3. Investigate fatality-related crashes.
4. Explore pedestrian, motorcycle, and matatu-related crashes.
5. Develop insights and recommendations for road safety.

## Key Questions

1. How have accidents changed over time?
2. Which months have the highest accident rates?
3. Which days of the week are most dangerous?
4. Which locations experience the most crashes?
5. What proportion of crashes involve fatalities?
6. How common are pedestrian, motorcycle, and matatu crashes?
