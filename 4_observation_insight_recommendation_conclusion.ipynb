# Observations, Insights, Recommendations & Conclusion

## Yearly Road Accident Trend Insights

The number of reported road crashes increased sharply between 2012 and 2015, rising from 310 incidents to a peak of 6,101 incidents in 2015. This period may reflect increased road usage, urbanization, growth in vehicle ownership, or improvements in crash reporting.

Following the 2015 peak, crash counts generally declined, although a temporary increase was observed in 2018. The overall downward trend after 2015 may indicate the impact of road safety initiatives, infrastructure improvements, or changes in reporting patterns.

A substantial decline in reported crashes was observed from 2019 onwards. This trend may partly reflect reduced mobility during the COVID-19 pandemic period and other transportation-related changes.

The low number of crashes recorded in 2023 should be interpreted cautiously until the completeness of the data for that year is confirmed.

## Seasonal Crash Pattern Insights

The monthly analysis revealed that road crashes occur throughout the year, although certain months experience higher crash frequencies than others.

March recorded the highest number of crashes (3,206), followed by July (3,030) and May (2,999). This suggests elevated road safety risks during the first half of the year, particularly between March and July.

A second period of increased crash activity was observed during November and December, which may be associated with increased holiday travel and traffic volumes.

Although April recorded the lowest number of crashes (2,552), the variation between months is relatively moderate, indicating that road accidents remain a persistent challenge throughout the year rather than being confined to a specific season.

These findings suggest that road safety interventions should be maintained year-round, with additional emphasis during March, the May–July period, and the end-of-year festive season.

## Quarterly Road Accident Analysis

### Observations

The quarterly analysis produced the following results:

| Quarter | Number of Crashes |
|----------|------------------:|
| Q1 (Jan–Mar) | 8,601 |
| Q2 (Apr–Jun) | 8,300 |
| Q3 (Jul–Sep) | 8,368 |
| Q4 (Oct–Dec) | 8,390 |

Q1 (January–March) recorded the highest number of crashes with 8,601 incidents, while Q2 (April–June) recorded the lowest with 8,300 incidents.

### Key Insights

#### Insight 1: The First Quarter Is the Highest-Risk Period

Q1 accounted for the highest number of crashes in the dataset. This suggests that road users may face elevated risks during the beginning of the year, particularly in March, which was also identified as the highest-risk month.

#### Insight 2: Crash Risk Remains Relatively Consistent Throughout the Year

The difference between the highest quarter (Q1) and the lowest quarter (Q2) is only 301 crashes. This suggests that road accidents are not heavily concentrated in a single season but persist as a challenge throughout the year.

#### Insight 3: Elevated Crash Activity at the End of the Year

Q4 recorded the second-highest number of crashes. This may be associated with increased travel, holiday activities, and higher traffic volumes during the festive season.

### Recommendations

Based on the quarterly analysis:

1. Road safety campaigns should be implemented throughout the year rather than focusing on a single season.
2. Additional enforcement and awareness programs should be considered during Q1, particularly in March, when crash frequencies are highest.
3. Traffic management and road safety measures should be strengthened during Q4 to address increased travel demand during the festive season.
4. Continuous monitoring of crash trends should be maintained to identify emerging seasonal patterns and support evidence-based road safety planning.

### Conclusion

The quarterly analysis indicates that road traffic crashes occur consistently throughout the year, with only moderate seasonal variation. While Q1 recorded the highest number of crashes, the relatively small differences across quarters suggest that road safety interventions should be sustained year-round, with targeted emphasis during high-risk periods such as the first quarter and festive season months.

## Day of Week Road Accident Analysis

### Observations

The distribution of road crashes across the days of the week is shown below:

| Day | Number of Crashes |
|------|------------------:|
| Monday | 4,976 |
| Tuesday | 5,113 |
| Wednesday | 5,008 |
| Thursday | 4,941 |
| Friday | 5,487 |
| Saturday | 4,738 |
| Sunday | 3,396 |

Friday recorded the highest number of crashes (5,487), while Sunday recorded the lowest (3,396).

### Key Insights

#### Insight 1: Friday Is the Highest-Risk Day

Friday experienced the highest number of crashes among all days of the week. This may be attributed to increased traffic volumes as people travel for work, business, social activities, and weekend-related journeys.

The combination of heavy traffic, driver fatigue accumulated during the work week, and increased evening travel may contribute to the elevated crash frequency.

#### Insight 2: Weekday Crash Levels Are Consistently High

From Monday to Friday, crash counts remain relatively stable, ranging between approximately 4,900 and 5,500 incidents.

This suggests that daily commuting and economic activities are major contributors to road crash occurrence throughout the working week.

#### Insight 3: Weekend Crash Patterns Differ

Crash frequencies decline during the weekend, particularly on Sunday.

Sunday recorded the lowest number of crashes in the dataset, which may reflect:

- Reduced commuter traffic
- Lower commercial transport activity
- Fewer vehicles on the road compared to weekdays

#### Insight 4: Economic Activity Drives Road Risk

The concentration of crashes during weekdays indicates a strong relationship between road usage intensity and crash occurrence. As traffic volumes increase during normal business operations, accident risk also increases.

### Recommendations

Based on the day-of-week analysis:

1. Traffic enforcement operations should be intensified on Fridays, particularly during evening hours.
2. Employers and transport operators should promote safe driving practices during peak commuting periods.
3. Road safety awareness campaigns should target weekday commuters and commercial vehicle operators.
4. Traffic management measures should focus on reducing congestion and improving road user behavior during high-traffic weekdays.

### Conclusion

The analysis reveals that road crashes occur more frequently during weekdays than weekends. Friday emerged as the highest-risk day, while Sunday recorded the lowest crash frequency. These findings suggest that commuter traffic and economic activities are key drivers of road accident occurrence and should be a primary focus of road safety interventions.

## Hourly Road Accident Analysis

### Observations

Road crash frequencies vary significantly throughout the day.

The highest number of crashes occurred during:

| Hour | Number of Crashes |
|--------|------------------:|
| 07:00 | 3,232 |
| 06:00 | 2,890 |
| 17:00 | 2,477 |
| 08:00 | 2,322 |
| 18:00 | 2,210 |

The lowest number of crashes occurred during the early morning hours between midnight and 5:00 AM.

### Key Insights

#### Insight 1: Morning Rush Hour Is the Most Dangerous Period

The highest crash frequency was recorded at 7:00 AM, with 3,232 crashes. Elevated crash levels were also observed at 6:00 AM and 8:00 AM.

This pattern strongly suggests that morning commuting periods represent the highest-risk time of day for road users.

Possible contributing factors include:

- Heavy traffic congestion
- High commuter volumes
- Time pressure and speeding
- Reduced driver attention during peak travel periods

#### Insight 2: Evening Rush Hour Also Presents Significant Risk

Crash frequencies increase again during the late afternoon and evening hours, peaking at 5:00 PM (17:00) and remaining high through 6:00 PM (18:00).

This second peak corresponds to the evening commute when large numbers of road users are returning home from work, school, and business activities.

#### Insight 3: Midday Crash Levels Are Relatively Stable

Between 10:00 AM and 4:00 PM, crash counts remain moderate and relatively stable compared to rush-hour periods.

This suggests that road risk is lower during non-peak travel hours.

#### Insight 4: Overnight Crash Activity Is Substantially Lower

The fewest crashes occur between 12:00 AM and 5:00 AM.

While these hours record the lowest crash counts, severe crashes may still occur due to factors such as:

- Reduced visibility
- Driver fatigue
- Impaired driving
- High-speed travel on less congested roads

### Recommendations

Based on the hourly analysis:

1. Traffic management efforts should prioritize morning rush-hour periods between 6:00 AM and 8:00 AM.
2. Additional enforcement and traffic control measures should be implemented during evening peak hours between 5:00 PM and 6:00 PM.
3. Public awareness campaigns should focus on commuter safety during peak travel periods.
4. Employers and transport operators should encourage safe driving practices during high-risk commuting hours.
5. Urban traffic planning should address congestion-related risks during morning and evening peak periods.

### Conclusion

The analysis reveals a clear bimodal pattern in road crash occurrence, with two major peaks occurring during morning and evening commuting hours. The highest risk period is between 6:00 AM and 8:00 AM, particularly at 7:00 AM, while crash frequencies decline substantially during overnight hours. These findings indicate that commuter traffic is a major contributor to road accident occurrence and should be a key focus of road safety interventions.

### Insight 5: Fatality Analysis (Fatality-Related Crashes)

Out of 33,659 recorded crashes, 2,450 were associated with fatality-related reports, representing 7.28% of all crashes. This means that approximately one in every fourteen crashes involved fatality-related keywords.

#### Key Findings

##### Fatality Burden

Fatality-related crashes accounted for a significant share of all reported crashes, highlighting the severity of road safety challenges.

##### Yearly Trend

Fatality-related crashes peaked in 2015 with 433 incidents and generally declined thereafter, although a temporary increase was observed in 2018.

##### Monthly Pattern

March recorded the highest number of fatality-related crashes (253), consistent with the overall crash trend observed during the monthly analysis.

##### Weekly Pattern

Fatality-related crashes were distributed relatively evenly across the week. Monday recorded the highest number of fatality-related crashes (371), while Thursday recorded the lowest (326).


## Fatality Rate Analysis

While the total number of fatality-related crashes provides insight into the overall burden of road fatalities, the fatality rate offers a better measure of crash severity by showing the proportion of crashes that involved fatalities.

### Fatality Rate by Year

The highest fatality rate was observed in 2023 (10.26%), followed by 2012 (9.68%) and 2020 (9.34%).

The lowest fatality rate occurred in 2016 (5.90%).

### Key Insights

#### Crash Severity Varies Across Years

The analysis reveals that years with fewer crashes do not necessarily have lower fatality risk. For example, although 2020 recorded substantially fewer crashes than 2015, a higher proportion of crashes were fatal.

#### 2016 Recorded the Lowest Fatality Risk

Crashes reported in 2016 were less likely to involve fatalities than in any other year in the dataset.

#### Recent Years Show Elevated Fatality Rates

Fatality rates increased in several recent years, including 2020, 2022, and 2023. This suggests that while crash frequencies may have declined, crash severity remains an important road safety concern.

### Recommendation

Road safety evaluations should consider both crash frequency and fatality rates. Reducing the number of crashes is important, but reducing crash severity and preventing fatalities should remain a primary objective of road safety interventions.

### Conclusion

Fatality-related crashes represent a smaller proportion of total crashes but remain a major public safety concern due to their severe consequences. The findings suggest that targeted interventions during high-risk periods and locations could help reduce the burden of road traffic fatalities.

#### Weekly Pattern

Friday recorded the highest number of pedestrian-related crashes (178), while Sunday recorded the lowest (118).

#### Hourly Pattern

Pedestrian crash risk peaks during both morning commuting hours and evening periods. The highest number of pedestrian crashes occurred at 7:00 AM (92 crashes), while elevated crash frequencies were also observed between 7:00 PM and 9:00 PM.

### Recommendations

- Enhance street lighting and visibility during evening hours.
- Increase pedestrian safety awareness campaigns.
- Prioritize interventions during peak pedestrian travel periods.

### Conclusion

Pedestrian-related crashes represent a smaller proportion of total crashes but highlight the vulnerability of non-motorized road users. The findings indicate that both commuting periods and evening hours require targeted pedestrian safety interventions.

## Matatu (Public Transport) Crash Analysis

### Observation

Out of 33,659 recorded crashes, 2,750 involved matatus, representing 8.17% of all reported crashes. This makes matatu-related crashes the largest special crash category identified in the analysis.

### Key Findings

#### Yearly Trend

Matatu-related crashes peaked in 2015 with 465 incidents and remained relatively high in 2018 with 431 incidents.

#### Monthly Pattern

Matatu crashes were distributed fairly evenly throughout the year, with February recording the highest number of crashes (260) and November recording the lowest (209).

#### Weekly Pattern

Wednesday recorded the highest number of matatu-related crashes (449), while Sunday recorded the lowest (258). Weekday crash frequencies were substantially higher than weekend crash frequencies.

#### Hourly Pattern

The highest number of matatu-related crashes occurred during morning commuting hours, particularly at 7:00 AM (316 crashes) and 6:00 AM (307 crashes). A secondary peak was observed during evening commuting hours.

### Recommendations

- Strengthen public transport safety programs and driver training initiatives.
- Increase enforcement during morning and evening rush-hour periods.
- Prioritize weekday public transport safety interventions.
- Promote compliance with speed limits and safe driving practices.

### Conclusion

Matatu-related crashes account for a significant share of road traffic incidents and exhibit strong associations with commuter travel patterns. The findings suggest that targeted public transport safety interventions could play an important role in reducing overall road crash risk.

## Geospatial Hotspot Analysis

### Observation

Geospatial analysis was conducted using latitude and longitude coordinates to identify areas with high concentrations of road crashes.

Interactive maps and hotspot analysis revealed that crashes are not evenly distributed across the road network but are concentrated in specific geographic locations.

### Key Findings

#### Crash Concentration

The most frequently reported crash hotspot recorded 201 crashes, indicating substantial spatial clustering of road accidents.

#### Corridor-Level Risk

Several hotspot coordinates were located in close proximity to one another, suggesting the presence of high-risk road corridors rather than isolated crash locations.

#### Urban Concentration

Most hotspot locations appear to be concentrated within the Nairobi metropolitan region, reflecting the area's high traffic density, economic activity, and transport demand.

### Implications

The findings suggest that targeted interventions at a relatively small number of high-risk locations could significantly reduce crash occurrence.

### Recommendations

- Conduct detailed road safety audits at identified hotspot locations.
- Improve road signage, lighting, and intersection design.
- Enhance pedestrian infrastructure in high-risk urban areas.
- Increase enforcement and traffic monitoring along major transport corridors.
- Prioritize hotspot locations for future road safety investments.

### Conclusion

Road crashes exhibit strong geographic clustering, with several locations accounting for a disproportionately large number of incidents. Targeted interventions at these hotspot locations offer significant opportunities for improving road safety outcomes.
