# Exploratory Data Analysis (EDA)

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
```

```python
# Load cleaned Dataset

df_accidents = pd.read_csv(
    "./data/processes/Kenya_road_accidents_cleaned.csv"
)

df_accidents.head()
```

## Question 1: How have crashes changed over time?

```python
# count crashes by year

crashes_by_year = df_accidents['year'].value_counts().sort_index()

crashes_by_year
```

```python
# visualize yearly trend

crashes_by_year.plot(
    kind='line',
    marker='s'
)

plt.title('Kenyan Road Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')


plt.savefig(
    "./images/crashes_by_year.png",
    bbox_inches='tight'
)

plt.show()
```

There was a steady increase in road accidents between 2015 and 2022, suggesting a growing traffic volume and increased reporting coverage as the years went by. 

## Question 2: Which months recorded the most crashes?

```python
month_order = [
    'January', 'February', 'March', 'April',
    'May', 'June', 'July', 'August',
    'September', 'October', 'November', 'December'
]
crashes_by_month = (
    df_accidents['month_name'].value_counts().reindex(month_order)
)

crashes_by_month
```

```python
# plot crashes by month

crashes_by_month.plot(
    kind = 'line',
    marker = 's'
)

plt.title('Kenyan Road accidents by Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')

plt.savefig(
    "./images/crashes_by_month.png",
    bbox_inches='tight'
)

plt.show()
```

## Question 3: Which days are most dangerous?

```python
weekday_order = [
    'Monday', 'Tuesday', 'Wednesday', 'Thursday',
    'Friday', 'Saturday', 'Sunday'
]
crashes_by_day = (
    df_accidents['day_name'].value_counts().reindex(weekday_order)
)

crashes_by_day
```

```python
# plot Weekday Trend

crashes_by_day.plot(
    kind='line',
    marker='s',
)

plt.xlabel('Days')
plt.ylabel('Number of Accidents')
plt.title('Kenyan Road Accidents by Day')

plt.savefig(
    "./images/crashes_by_day.png",
    bbox_inches='tight'
)

plt.show()
```

## Question 4: Which monthly quarter_period were the highest?

```python

crashes_by_quarter = df_accidents['quarter_period'].value_counts()

crashes_by_quarter
```

```python
# plot hourly trend

crashes_by_quarter.plot(
    kind='line',
    marker='s'
)

plt.xlabel('Monthly Quarterly Period')
plt.ylabel('Number of Accidents')
plt.title('Kenyan Road Accidents per Quarter')

plt.savefig(
    "./images/crashes_by_quarter.png",
    bbox_inches='tight'
)

plt.show()
```

## Question 4: Which hours are riskiest?

```python

crashes_by_hour = df_accidents['hour'].value_counts().sort_index()

crashes_by_hour
```

```python
# plot hourly trend

crashes_by_hour.plot(
    kind='line',
    marker='s'
)

plt.xlabel('Time')
plt.ylabel('Number of Accidents')
plt.title('Kenyan Road Accidents Hourly')

plt.savefig(
    "./images/crashes_by_hour.png",
    bbox_inches='tight'
)

plt.show()
```

## Question 5: What proportion of crashes involve fatalities?

```python
df_accidents['contains_fatality_words'].value_counts()
```

```python
# create a pie chart
Fatality_pie = df_accidents['contains_fatality_words'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel('')
plt.title('Fatality-Related Crashes vs Non-Fatality Crashes')

plt.savefig(
    "./images/Fatality-Related Crashes vs Non-Fatality Crashes.png",
    bbox_inches='tight'
)        

plt.show()
```

```python
# percentage calculation
fatality_pct = df_accidents['contains_fatality_words'].mean()*100

fatality_pct
```

```python
# fatalities by year
fatalities_by_year = df_accidents[df_accidents['contains_fatality_words'] == True].groupby('year').size()

fatalities_by_year
    
```

```python
# fatalities by month
fatalities_by_month = df_accidents[
    df_accidents['contains_fatality_words'] == True
].groupby('month_name').size()

fatalities_by_month
```

```python
# fatalities by day
fatalities_by_day = df_accidents[
    df_accidents['contains_fatality_words'] == True
].groupby('day_name').size()

fatalities_by_day
```

```python
# fatality rate by year
fatality_rate_by_year = (
    df_accidents.groupby('year')['contains_fatality_words']
    .mean()
    * 100
)

fatality_rate_by_year
```

```python
# visualization of fatality rate by year
fatality_rate_by_year.plot(
    kind= 'line',
    marker= 's'
)

plt.title('Fatality Rate by Year')
plt.xlabel('Year')
plt.ylabel('Fatality Rate (%)')

plt.savefig(
    "./images/Fatality Rate by Year (%).png",
    bbox_inches='tight'
)

plt.show()
```

## Question 7: Which locations experience the most crashes?

```python
# find center of the data

center_latitude = df_accidents['latitude'].mean()

center_longitude = df_accidents['longitude'].mean()

center_latitude, center_longitude
```

Folium must have a starting point and average of the coordinates place the map near Nairobi, where most crash reports are concentrated.

```python
# create a base map
kenyan_map = folium.Map(
    location=[center_latitude, center_longitude],
    zoom_start = 11
)
```

```python
# display the map

kenyan_map
```

```python
# add accident markers

sample_accidents_df = df_accidents.sample(
    150,
    random_state=42
)
sample_accidents_df.head()
```

```python
for _, row in sample_accidents_df.iterrows():

    folium.CircleMarker(
        location=[
            row['latitude'],
            row['longitude']
        ],
        radius=3,
        fill=True
    ).add_to(kenyan_map)


kenyan_map
```

```python
# creating a heat map
heatmap_data = df_accidents[
    ['latitude', 'longitude']
].values.tolist()
```

```python
# create heat map
kenyan_heatmap = folium.Map(
    location=[
        center_latitude,
        center_longitude
    ],
    zoom_start=11
)
```

```python
# add heat layer
HeatMap(
    heatmap_data
).add_to(
    kenyan_heatmap
)
```

```python
kenyan_heatmap
```

Density visualization is displayed.

 - Red = High concentration of road crashes
 - Yellow = Medium or moderate concentration of road crashes
 - Blue = few or low concentration of road accidents

This helps reveal hotspots

```python
# save the map
kenyan_heatmap.save(
    "./reports/kenya_accident_heatmap.html"
)
```

```python
# identify top 30 hotspots

hotspots_top30 = (
    df_accidents.groupby(['latitude','longitude'])
    .size().reset_index(name='crash_count')
    .sort_values('crash_count', ascending=False)
)

hotspots_top30.head(30)
```

```python
# create hotspots markers
hotspot_kenyan_map = folium.Map(
    location=[
        center_latitude,
        center_longitude
    ],
    zoom_start=11
)
```

```python
for _, row in hotspots_top30.head(30).iterrows():

    folium.Marker(
        location=[
            row['latitude'],
            row['longitude']
        ],
        popup=f"Crashes: {row['crash_count']}"
    ).add_to(
        hotspot_kenyan_map
    )
```

```python
hotspot_kenyan_map
```

```python
# save the hotspot map
hotspot_kenyan_map.save(
 "./reports/top_30_hotspots_map"
)
```

```python
# extract the top 10 hotspots
top_hotspots = hotspots_top30.head(10)

top_hotspots
```

### Geospatial Analysis Findings

The heatmap reveals that road traffic crashes are highly concentrated in specific urban corridors and major transportation routes. Several hotspots exhibit significantly higher crash densities than surrounding areas, suggesting the need for targeted road safety interventions, traffic management improvements, and infrastructure upgrades.
