## Dataset Description

Source: Kenya Road Traffic Crashes Dataset (2012–2023)

Variables in both datasets:

- crash_id: Unique crash identifier
- crash_datetime: Date and time of crash
- crash_date: Date of crash
- latitude: Crash latitude
- longitude: Crash longitude
- n_crash_reports: Number of reports
- contains_fatality_words: Indicates fatality-related crash
- contains_pedestrian_words: Indicates pedestrian involvement
- contains_matatu_words: Indicates matatu involvement
- contains_motorcycle_words: Indicates motorcycle involvement

### Considering the column names match, I will have to combine the datasets

```python
krtc_df = pd.concat([alg_df,manual_df], ignore_index=True)

krtc_df
```
