# Data Cleaning and Preparation

```python
# create a working copy
df_accidents = crashes_df.copy()
```

I have done this to avoid messing up or damaging the original dataset i.e crashes_df

```python
# check the data types

df_accidents.dtypes
```

crash_datetime and crash_date columns have to be converted to a date format from text or string format so that it can be easier to extract
- Year
- Month
- Day
- Hour

This will help us answer questions in our objectives such as what year recorded the highest number of crashes, Identify high risk months many other questions.

```python
# Convert date columns
df_accidents['crash_datetime'] = pd.to_datetime(df_accidents['crash_datetime'])

df_accidents['crash_date'] = pd.to_datetime(df_accidents['crash_date'])
```

```python
# verify the conversions
df_accidents[['crash_datetime', 'crash_date']].dtypes
```

```python
# create year column
df_accidents['year'] = df_accidents['crash_datetime'].dt.year

# create month column
df_accidents['month'] = df_accidents['crash_datetime'].dt.month

df_accidents
```

```python
# create the month name

df_accidents['month_name'] = df_accidents['crash_datetime'].dt.month_name()
df_accidents
```

The months have been named as either January, February and so on rather than 1st, 2nd.

```python
# create day of the week
df_accidents['day_name'] = (df_accidents['crash_datetime'].dt.day_name())

df_accidents
```

This will help in knowing what days had the most accidents. 

```python
# create hour
df_accidents['hour'] = df_accidents['crash_datetime'].dt.hour

df_accidents.head()
```

These variables will assist in identifying morning rush hour, afternoon traffic, evening congestion and latenight accidents

```python
# check missing values again
df_accidents.isnull().sum()
```

All records were perfectly converted

```python
# inspect missing coordinates
df_accidents[df_accidents['latitude'].isnull() |
    df_accidents['longitude'].isnull()
    ]
```

Coordinates will assist in hotspot analysis that's why it is important to know how many are missing and if the number is significant
